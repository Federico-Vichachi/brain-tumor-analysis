```python
numeric_cols = ["Age", "Survival_Rate", "Tumor_Growth_Rate", "Tumor_Size"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```python
def calculate_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
numeric_columns_1 = numeric_cols[:2]
numeric_columns_2 = numeric_cols[2:]
```

```python
for col in numeric_cols:
    outliers = calculate_outliers(col)
    print(f"Outliers en {col}: {len(outliers)}")
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_columns_1], orient="h", palette="Set2", showfliers=True)
plt.title("Boxplots para verificar la presencia de outliers")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_columns_2], orient="h", palette="Set2", showfliers=True)
plt.title("Boxplots para verificar la presencia de outliers")
plt.show()
```

```python
df.set_index("Patient_ID", inplace=True)
df.head()
```

```python
df_encoded = df.copy()
```

```python
gender_map = {'Male': 1, 'Female': 0}
yes_no_map = {'Yes': 1, 'No': 0}
mri_result_map = {'Positive': 1, 'Negative': 0}
stage_map = {'I': 1, "II": 2, "III": 3, "IV": 4}
```

```python
df_encoded["Gender"] = df_encoded["Gender"].map(gender_map)
df_encoded["MRI_Result"] = df_encoded["MRI_Result"].map(mri_result_map)
df_encoded["Stage"] = df_encoded["Stage"].map(stage_map)
```

```python
columns_to_map = ["Radiation_Treatment", "Surgery_Performed", "Chemotherapy", "Family_History", "Follow_Up_Required"]
```

```python
for col in columns_to_map:
    df_encoded[col] = df_encoded[col].map(yes_no_map)
```
