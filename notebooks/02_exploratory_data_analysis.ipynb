```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/brain_tumor_dataset.csv', index_col=0)
```

```python
df.head()
```

```python
categorical_vars = ['Gender', 'Tumor_Type', 'Location', 'Histology', 'Stage', 
                    'Radiation_Treatment', 'Surgery_Performed', 'Chemotherapy', 
                    'Family_History', 'MRI_Result']


for var in categorical_vars:
    plt.figure(figsize=(8, 5))
    x = var
    sns.countplot(data=df, x=x, palette='deep', hue=x, order=df[var].value_counts().index)
    plt.title(f'Distribucin de la variable {var}')
    plt.xlabel(var)
    plt.ylabel('Cantidad')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
sns.barplot(data=df, x='Gender', y='Tumor_Size', palette='deep', errorbar=None)
plt.title('Tamao del tumor segn el gnero')
plt.show()
```

```python
sns.barplot(data=df, x='Stage', y='Tumor_Size',hue='Tumor_Type', errorbar=None, palette='colorblind')
plt.legend(title='Tipo de Tumor', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.countplot(data=df, x='Histology', hue='Tumor_Type', palette='bright')
plt.legend(title='Tipo de Tumor', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.barplot(data=df, x='Location', y='Tumor_Size', errorbar=None, palette='pastel')
plt.title('Tamao del tumor segun su localizacin cerebral')
plt.show()
```

```python
stage_map = {'I': 1, "II": 2, "III": 3, "IV": 4}
df["Stage"] = df["Stage"].map(stage_map)
```

```python
sns.countplot(data=df, x='Symptom_1', hue='Gender')
plt.title('Conteo del primer sintoma observado en hombres y mujeres')
plt.show()
```
