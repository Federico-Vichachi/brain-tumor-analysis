# 📂 02 - Exploratory Data Analysis

## 0. Introducción 🤔❓
En este notebook realizamos el Análisis Exploratorio de Datos (EDA) del conjunto que simula información clínica sobre tumores cerebrales.  
El objetivo es comprender la distribución de variables, detectar patrones y visualizar posibles relaciones entre características que serán claves para la presentación.

## 1. Importamos las librerías necesarias 📚📊

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Carga de Datos 📃


```python
df = pd.read_csv('../data/brain_tumor_dataset.csv', index_col=0)
df_encoded = pd.read_csv('../data/brain_tumor_encoded_dataset.csv', index_col=0)
```

## 3. Definimos la paleta de colores que utilizaremos en los gráficos 🎨

```python
palette = ['#3F3038', '#FDAED8','#854836', '#F361AF',]
bg_color = '#FDF6FA'
```

## 4. Distribución de variables categóricas 📊
Se utiliza countplot para observar la frecuencia de cada categoría en variables como género, tipo de tumor, localización, tipo de histología, entre otras.

```python
categorical_vars = ['Gender', 'Tumor_Type', 'Location', 'Histology', 'Stage', 
                    'Radiation_Treatment', 'Surgery_Performed', 'Chemotherapy', 
                    'Family_History', 'MRI_Result']
```

```python
for var in categorical_vars:
    plt.figure(figsize=(8, 5))
    x = var
    sns.countplot(data=df, x=x, palette=palette, hue=x, order=df[var].value_counts().index)
    plt.title(f'Distribución de la variable {var}')
    plt.gcf().patch.set_facecolor(bg_color)  
    plt.xlabel(var)
    plt.ylabel('Cantidad')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

## 5. Gráficos para responder las preguntas y temas de interés de la presentación 🧠

```python
sns.barplot(data=df, x='Gender', y='Tumor_Size', palette=palette, errorbar=None)
plt.title('Tamaño del tumor según el género')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.barplot(data=df, x='Stage', y='Tumor_Size',hue='Tumor_Type', errorbar=None, palette=palette)
plt.legend(title='Tipo de Tumor', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gcf().patch.set_facecolor(bg_color)  
plt.show()
```

```python
sns.countplot(data=df, x='Histology', hue='Tumor_Type', palette=palette)
plt.legend(title='Tipo de Tumor', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gcf().patch.set_facecolor(bg_color)
plt.show()
```

```python
sns.barplot(data=df, x='Location', y='Tumor_Size', errorbar=None, palette=palette)
plt.title('Tamaño del tumor segun su localización cerebral')
plt.gcf().patch.set_facecolor(bg_color)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.countplot(data=df, x='Symptom_1', hue='Gender', palette=palette)
plt.title('Conteo del primer sintoma observado en hombres y mujeres')
plt.legend(title='Tipo de Tumor', loc='upper left', bbox_to_anchor=(1, 1))
plt.gcf().patch.set_facecolor(bg_color)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.boxplot(data=df,x='Histology', y='Survival_Rate', palette=palette)
plt.gcf().patch.set_facecolor(bg_color)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.boxplot(data=df,x='Location', y='Survival_Rate', palette=palette)
plt.gcf().patch.set_facecolor(bg_color)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.boxplot(x=df['Chemotherapy'], y=df_encoded['Stage'], palette=palette)
plt.gcf().patch.set_facecolor(bg_color)
plt.yticks(range(df_encoded['Stage'].min(), df_encoded['Stage'].max() + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 6. Heatmap para verificar correlación entre variables númericas 🌡🔥

```python
# 7. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor para identificar correlaciones')
plt.show()
```
